## Part 1: summation 

Your algorithms may not use `sum` or `np.sum` or `math.fsum` (except for the *stoppping rule* given for `fan`), but you will benchmark your algorithms against these functions to observe their relative **numerical error**.

For the `"sorted"` and `"kahan"` functionalities, implement the `return_cumulative_sequence` argument which returns the full sequence of all cumulative summations rather than just the final total sum if set to `True`.

```python
def summation(x, method="kahan", *args, **kwargs):
    """
    x                : (list) [x[0], x[1], ..., x[n-1]] of numbers
    method(="kahan") : (str)  <'sorted'|'fan1'|'fan2'|'kahan'> 
    """
``` 

  
- `"sorted"` assumes `x` is already sorted by increasing magnitude and adds values in the order given sequentially (but the sorting will actually need to be done externally to this function)
 
 
- `"fan"` methods assume the values in `x` are of roughly equal magnitude and deploy "divide and conquer" algorithms to sum the values of `x` in a "pairwise" manner
    - `"fan1"` sums the even and odd pairs `x[0::2]` "+" `x[1::2]` appending an extra 0 onto the odds if needed, and recursively calls itself on the "halved" list
    - `"fan2"` splits the list "in two" and returns the sum of two recursive calls to itself, one for `x[:int(len(x)/2)]` and the other for `x[int(len(x)/2):]`
    - All `"fan"` methods use the same *stopping rule*  and return `sum(x)` if `len(x) <= 3`  


- `"kahan"` sums the numbers acording to [Kahan's improved summation algorithm](https://en.wikipedia.org/wiki/Kahan_summation_algorithm) designed to sequentially minimize accumulated ***roundoff error*** $\big|\sum_{i=1}^k x_i - \left[\sum_{i=1}^k x_i\right]_c\big|$ of the *computer representation* $[\cdot]_c$: 


> $\begin{align*} & {} s = x_1 & {} & \texttt{ # initialize sequential sum}\\ & {} e = 0 & {} & \texttt{ # roundoff error adjustment}\\ & {} \text{for } i = 2, ..., n \\ & {} \{ \\ & {} \quad \quad \tilde x = x_i - e & {} &  \texttt{ # previous roundoff error adjustment}\\ & {} \quad \quad \tilde s = s + \tilde x & {} & \texttt{ # sum at $i$}\\ & {} \quad \quad e = (\tilde s - s) - \tilde x & {} & \texttt{ # detectable roundoff error in } s+\tilde x\\ & {} \quad \quad s = \tilde s & {} & \texttt{ # sum so far}\\ & {} \} \end{align*}$

> *This problem is inspired by the **Addition of Several Numbers** section in Chapter 2.3 **Computer Storage and Arithmatic: Errors** following Chapter 2.2 **The Floating-Point Number System** on page 99 of James E. Gentle's **Computational Statistics** textbook.*    
    


````python
# this might have been called `sequential_sum` but naming it
# `sorted_sum` emphasizes that we will have interest in a sorted sum
def sorted_sum(x, return_cumulative_sequence=False):
    
    """
    Sequential summation of x in the order given  
    *assuming x is already sorted* by increasing magnitude
    
    x : (list) [x[0], x[0], ..., x[n-1]] of numbers
    return_cumulative_sequence(=False) : return the full sequence of cumulative sums
                                         rather than the final summation value if True
                                         
    return <float|list>
    """   
        
    cumulative_sum = list(x)
    for i,xi in enumerate(x[1:]):
        pass
        
    if return_cumulative_sequence:
        return cumulative_sum
    return cumulative_sum[-1]
        

def fan1(x):   
    
    """
    Pairwise "fan" summation of x based on even-odd divide and conquer
    
    x : (list) [x[0], x[0], ..., x[n-1]] 
               are assumed to be of roughly the same magnitude
    
    returns (float) after recursive calls
    """

    if len(x) <= 3:
        return sum(x)
    
    # convert to np.array to sum
    # then convert back to list 
    
    return fan1(x[:int(round(len(x)/2))])


def fan2(x):   
    
    """
    Pairwise "fan" summation of x based on recursive divide and conquer
    
    x : (list) [x[0], x[0], ..., x[n-1]] 
               are assumed to be of roughly the same magnitude
    
    returns (float) after recursive calls
    """

    if len(x) <= 3:
        return sum(x) 
    
    # just fix these and that's it...
    return fan2(x)+fan2(x)


def kahan(x, return_cumulative_sequence=False):
    
    """
    Summation of x using the Kahan algorithm
    
    x : (list) [x[0], x[0], ..., x[n-1]] of numbers
    return_cumulative_sequence(=False) : return the full sequence of cumulative sums
                                         rather than the final summation value if True    
                                         
    return <float|list>                                         
    """
    
    cumulative_sum = list(x)
    
    for i,xi in enumerate(x[1:]):
        pass 
    
    if return_cumulative_sequence:            
        return cumulative_sum
    return cumulative_sum[-1]


# The `**kwargs` construct allows any optional argument to be passed to the `summation` function
# https://stackoverflow.com/questions/9539921/how-do-i-create-a-python-function-with-optional-arguments
def summation(x, method="kahan", *args, **kwargs):

    if method=='sorted':
        
        if 'return_cumulative_sequence' in kwargs:
            arg_to_pass_forward = kwargs['return_cumulative_sequence']
            return sorted_sum(x, return_cumulative_sequence=arg_to_pass_forward)        
        return sorted_sum(x)
    
    elif method=='fan1':       
        return fan1(x)
    elif method=='fan2':       
        return fan2(x)
    
    else:
        
        if 'return_cumulative_sequence' in kwargs:
            arg_to_pass_forward = kwargs['return_cumulative_sequence']
            return kahan(x, return_cumulative_sequence=arg_to_pass_forward)        
        return kahan(x)    
````

All numerical methods will be contrained by the limitations of the floating-point number system ${\rm I\!F}$.

Here is a ***bisection method*** for numerically finding the ***roots*** of *continuous* functions $f(x)$ (and ***critical points*** of *continuous* functions $g(x)$ if $f(x) = g'(x)$].

````python
# Look at the code below first before looking at
# the details of this function if you are intersted 
def bisection_root_finder(f, a, b, steps, ax=None):
    """
    Find a root of a function with the bisection method
    f: target function
    a: initial left bound
    b: initial right bound

    f(a) and f(b) must have different signs so we know they bound 0!
    """

    if ax:
        f_min = np.min([f(a),f(b)])
        f_max = np.max([f(a),f(b)])
        xs = np.linspace(a, b, 100) 
        ax.plot([a,b], 2*[0], 'k--')
        ax.plot(xs, f(xs))

    for i in range(steps):

        if ax:
            ax.plot([a,b], 2*[f_max - (i-1)*(f_max-f_min)/(steps-1)], 'k')
            ax.text(a, f_max - (i-1)*(f_max-f_min)/(steps-1), '|', ha='right', va='center')
            ax.text(b, f_max - (i-1)*(f_max-f_min)/(steps-1), '|', ha='left', va='center')

        if np.sign(f(a)) == 0:
            return a,f(a)
        elif np.sign(f(b)) == 0:
            return b,f(b)
        elif np.sign(f(a)) == np.sign(f(b)):
            print("Something Unexpected Happened")
        else:
            if np.sign(f(a)) == np.sign(f(a + (b-a)/2)): 
                a = a + (b-a)/2 
                # this is potentially better than (a+b)/2 since `a` being 
                # closer to `(b-a)/2` than `b` could improve numerical accuray
            else:
                b = b - (b-a)/2
                # this one doesn't matter in the same way
                
    return a+(b-a)/2, f(a+(b-a)/2)
````

## Part 3.B: Taylor Series Expansions 

--- 

The ***Taylor Series*** approximation of $f(x)$ expanded around $x_0$ is

$$f(x) = \underset{f^{(k)} \text{ is the } k^{th} \text{derivative of } f}{\sum_{k=0}^{\infty} \frac{f^{(k)}(x_0)}{k!}(x-x_0)^k}$$

For example, the ***Taylor Series*** expanded around $0$ of 
$\quad \ln(1+x) = x - \frac{x^2}{2} + \frac{x^3}{3} + \cdots \quad$ and for $e^x$ it's

$$e^x  = e^0 + e^0(x-0) + \frac{e^0}{2}(x-0)^2 + \frac{e^0}{3!}(x-0)^3 + \cdots = 1 + x + \frac{x^2}{2} + \frac{x^3}{3!} + \cdots $$



<!-- > *The examples in this section are is taken from Keith Knight's STA410 [notes1.pdf](https://q.utoronto.ca/courses/296804/files?preview=24222782) document*. -->


````python
def taylors_e(x, approximation_limit=100, verbose=False):
    
    """
    Taylor Series Approximation of e^x using 
    approximation_limit (default 100) terms
    """ 
    
    xi, ifac, ex =  1., 1., 1.
    series = np.zeros(approximation_limit)

    for i in range(approximation_limit):
        xi *= x
        ifac *= (i+1)
        ex += xi/ifac
        if np.isinf(xi) or np.isinf(ifac):
            print("Edge of the floating point universe encountered")
        if verbose:
            series[i] = ex
    if verbose:
        return series
    return ex 
````

````python
x = 50
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(15,4))
ax[0].plot(taylors_e(-x, approximation_limit=100, verbose=True), 
           label='$e^{-'+str(x)+'}$')
ax[0].set_xlabel("Rapidly oscilating cumulative sum")
ax[0].legend()
ax[1].plot(taylors_e(x, approximation_limit=100, verbose=True), 
           label='$e^{'+str(x)+'}$');
ax[1].set_xlabel("Numer of Terms in Taylor Series Approximation")
ax[1].legend()
ax[2].plot([0,99], 2*[-x], label='Truth')
ax[2].plot(np.log(1./taylors_e(x, approximation_limit=100, verbose=True)), 
           label='$\\log(1/e^{'+str(x)+'})$')
ax[2].set_xlabel("Smooth numerical convergence to truth")
ax[2].legend();
````
